```python
def calculate_grade(marks):
    """
    Returns the grade based on marks obtained.
    
    Parameters:
        marks (int or float): The marks scored
        
    Returns:
        str: The corresponding grade
    """
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"


# Example usage:
print(calculate_grade(95))   # Output: A
print(calculate_grade(86))   # Output: B
print(calculate_grade(74))   # Output: C
print(calculate_grade(69))   # Output: Fail
print(calculate_grade(90))   # Output: A
print(calculate_grade(45))   # Output: Fail
```

```python
# 1. Even or Odd Number Checker
def is_even(number):
    """
    Returns True if the number is even, False if odd.
    Works with integers and floats (checks the integer part).
    """
    return number % 2 == 0


# 2. Factorial using a loop
def factorial(n):
    """
    Returns the factorial of a non-negative integer n.
    Returns 1 for 0 and 1.
    """
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result


# 3. Find maximum among three numbers
def find_max(a, b, c):
    """
    Returns the largest number among a, b, and c.
    """
    return max(a, b, c)
    # Alternative without using max():
    # if a >= b and a >= c:
    #     return a
    # elif b >= c:
    #     return b
    # else:
    #     return c


# 4. Word counter in a sentence
def count_words(sentence):
    """
    Returns the number of words in the given sentence.
    Splits on whitespace and ignores extra spaces.
    """
    if not sentence:
        return 0
    
    # split() without arguments splits on any whitespace and removes extra spaces
    words = sentence.split()
    return len(words)


print("is_even(4)   →", is_even(4))      # True
print("is_even(7)   →", is_even(7))      # False
print("is_even(0)   →", is_even(0))      # True
print("is_even(-6)  →", is_even(-6))     # True

print("\nfactorial(5)  →", factorial(5))    # 120
print("factorial(0)   →", factorial(0))    # 1
print("factorial(1)   →", factorial(1))    # 1
# print(factorial(-3))  # would raise ValueError

print("\nfind_max(10, 45, 23)  →", find_max(10, 45, 23))   # 45
print("find_max(7, 7, 7)      →", find_max(7, 7, 7))       # 7
print("find_max(-5, 0, -10)   →", find_max(-5, 0, -10))    # 0

print('\ncount_words("I love Python programming")          →', 
      count_words("I love Python programming"))            # 4
print('count_words("   hello   world  ")                   →', 
      count_words("   hello   world  "))                   # 2
print('count_words("Python")                               →', 
      count_words("Python"))                               # 1
print('count_words("")                                     →', 
      count_words(""))                                     # 0
```
